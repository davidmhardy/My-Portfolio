```python
# Get excess return
# Merge returns and risk-free rate
returns = pd.read_csv("MF_Data.csv")
returns
```

```python
returns['Date'] = pd.to_datetime(returns.Date)
returns['Date'] = returns.Date.dt.strftime('%Y-%m')
returns.set_index('Date', inplace=True)
returns
```

```python
factors = pd.read_csv('ff3_factors.csv')
```

```python
factors
```

```python
rf = factors[['Date','RF']]
rf
```

```python
rf.set_index('Date', inplace=True)
```

```python
rf
```

```python
# Substract risk-free rate (rf) from net returns of each mutual fund (Complete this step)
excess_returns = returns. sub(rf['RF'], axis= 'index')
excess_returns
```

```python
#Save excess returns
excess_returns.to_csv("MF_Excess_Returns.csv")
```

```python
mean_ex_ret = excess_returns.mean()
mean_ex_ret
```

```python
mean_ex_ret.to_csv("mean_ex_ret.csv")
```

```python
performance = pd.read_csv("Summary Stat.csv")
performance 
```

```python
performance.rename(columns={'Unnamed: 0':'fund'}, inplace=True)
performance
```

```python
mean_ex_ret = pd.read_csv("mean_ex_ret.csv")
mean_ex_ret
```

```python
mean_ex_ret.rename(columns={'Unnamed: 0':'fund','0': 'mean ex_ret'}, inplace=True)
mean_ex_ret
```

```python
# Merge previous computed performance measures and the excess returns (Complete this step)
performance_sharpe = pd.merge(performance, mean_ex_ret, on ="fund")
performance_sharpe
```

```python
# Compute Sharpe Ratio for each fund (Complete this step)
performance_sharpe['Sharpe Ratio'] = performance_sharpe['mean ex_ret']/performance_sharpe['std']
```

```python
performance_sharpe
```

```python
performance_sharpe = performance_sharpe.drop(columns = ['mean ex_ret'])
performance_sharpe
```

### Compute CAPM alpha and beta for each fund

```python
all_y = pd.read_csv('MF_Excess_Returns.csv', index_col = "Date") 
```

```python
all_y
```

```python
factors = pd.read_csv('ff3_factors.csv')
factors
```

```python
all_x = factors[['Date','Mkt-RF']]
all_x
```

```python
# Run regressions (Complete this step)
regression_results = all_y.apply(lambda x: linregress(all_x['Mkt-RF'], x), result_type='expand').rename(index={0:'slope',1:'intercept',2: 'rvalue', 3:'pvalue', 4:'stderr'})
regression_results                                                                                                               
```

```python
regression_results = regression_results.transpose()
regression_results
```

```python
regression_results['r-squared']=regression_results['rvalue']**2
regression_results
```

```python
regression_results.to_csv('CAPM alpha and beta.csv')
```

## Part 4: Form a portfolio

```python
fund_returns
```

```python
# Get equal-weight portfolio returns
weight = np.empty(len(selected_funds))
weight.fill(1/len(selected_funds))
weight
```

```python
np.sum(weight)
```

```python
# Compute portfolio returns (Complete this step)
portfolio = fund_returns.dot(weight)
portfolio
```

```python
# Get market returns for the same period
```
